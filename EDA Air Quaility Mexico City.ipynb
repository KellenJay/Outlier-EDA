# 3. Outlier Analysis
**Detect Outliers (IQR Method):** Create a reusable function for detecting outliers.

```python
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
```

**Analyze Outliers for Each Pollutant:** Store and summarize the outliers for later analysis.

```python
outliers_summary = {}
for pollutant in pollutants:
    outliers = detect_outliers(df, pollutant)
    outliers_summary[pollutant] = len(outliers)
    print(f'{pollutant}: {len(outliers)} outliers detected')
```

**Extract Outliers:** Isolate the outliers in the data using the previously calculated bounds.

```python
# Calculate IQR and bounds for SO2_MER
Q1_SO2 = df['SO2_MER'].quantile(0.25)
Q3_SO2 = df['SO2_MER'].quantile(0.75)
IQR_SO2 = Q3_SO2 - Q1_SO2
lower_bound_SO2 = Q1_SO2 - 1.5 * IQR_SO2
upper_bound_SO2 = Q3_SO2 + 1.5 * IQR_SO2

# Calculate IQR and bounds for CO_MER
Q1_CO = df['CO_MER'].quantile(0.25)
Q3_CO = df['CO_MER'].quantile(0.75)
IQR_CO = Q3_CO - Q1_CO
lower_bound_CO = Q1_CO - 1.5 * IQR_CO
upper_bound_CO = Q3_CO + 1.5 * IQR_CO

# Extract outliers
so2_outliers = df[(df['SO2_MER'] > upper_bound_SO2) | (df['SO2_MER'] < lower_bound_SO2)]
co_outliers = df[(df['CO_MER'] > upper_bound_CO) | (df['CO_MER'] < lower_bound_CO)]

# Display counts of outliers
print(f"Number of SO2_MER outliers: {len(so2_outliers)}")
print(f"Number of CO_MER outliers: {len(co_outliers)}")

```
